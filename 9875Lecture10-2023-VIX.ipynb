### Valuation of volatility derivatives given vanilla prices

- In Lecture 5, we saw that if volatility and stock price moves are
uncorrelated, volatility derivatives may in principle be valued given
just the prices of European options on stock.

    - The zero-correlation assumption is completely unrealistic but the
connection between European option prices and volatility derivative
values is suggestive.


- How tightly do the prices of European options (which are assumed to be
known) constrain the fair value of a volatility derivative?

### A simple lognormal model
$$
\newcommand{\E}{\mathbb{E} }
$$

Assume that $\log\left(\sqrt{\langle x \rangle_T}\right)$ is normally
distributed with mean $\mu$ and variance $s^2$. Then
$\log\left({\langle x \rangle_T}\right)$ is also normally distributed
with mean $2 \,\mu$ and variance $4\,s^2$. 

Volatility and variance swap
values are given by respectively

$${\mathbb{E} }[ {\sqrt {\left\langle x \right\rangle _T } } ] = e^{\mu + s^2
/2} ;\,{\mathbb{E} }\left[ {\left\langle x \right\rangle _T} \right] = e^{2\mu
+ 2s^2 }.$$ 

Solving for $\mu$ and $s^2$ gives

$$s^2  = 2\log \left( {\frac{{\sqrt {{\mathbb{E} }\left[ {\left\langle X
\right\rangle _T } \right]} }}{{{\mathbb{E} }\left[ {\sqrt {\left\langle X
\right\rangle _T } } \right]}}} \right);\,\,\mu  = \log \left(
{\frac{{{\mathbb{E} }\left[ {\sqrt {\left\langle X \right\rangle _T } }
\right]^2 }}{{\sqrt {{\mathbb{E} }\left[ {\left\langle X \right\rangle _T }
\right]} }}} \right)$$ 

and the convexity adjustment is given by


$$\sqrt {{\mathbb{E} }\left[ {\left\langle X \right\rangle _T } \right]}  -
{\mathbb{E} }\left[ {\sqrt {\left\langle X \right\rangle _T } } \right] =
\left( {e^{s^2 /2}  - 1} \right)\,{\mathbb{E} }\left[ {\sqrt {\left\langle X
\right\rangle _T } } \right].
$$

### Black-Scholes-style formula

Under this lognormal assumption, calls on variance may be valued using a
Black-Scholes style formula:

<a name="eq:BSvarcall"></a>(1)
$$\mathbb{E}\left[ \langle x \rangle_T -K\right]
^{+}=e^{2\mu
+2s^{2}}N(\tilde{d}_{+})-K\,N(\tilde{d}_{-})$$

with

\begin{eqnarray*}
\tilde{d}_{+} &=&\frac{-\frac{1}{2}\log K+\mu +2s^{2}}{s} \\
\tilde{d}_{-} &=&\frac{-\frac{1}{2}\log K+\mu }{s}.
\end{eqnarray*}

- <span>This simple lognormal assumption seems not too unreasonable</span>

    - <span>The empirical distribution of implied volatility changes looks
lognormal.</span>

    - <span>Also, the dynamics of the volatility skew are consistent with
approximately lognormal volatility dynamics.</span>


- <span>In this model, the variance and volatility swap values (or
equivalently the variance swap value plus the convexity adjustment) are
all that is required to fix the values of all options on variance.
</span>

- <span> Moreover, at least for the major equity indices, there is a tight
market in variance swaps and a somewhat less liquid market in the
convexity adjustment.</span>

### A Heston example

Suppose the true dynamics of the underlying were Heston (recall that
lognormal volatility is a much more realistic assumption) with BCC
parameters:

$$\lambda = 1.15,\rho =0,\sigma _{0}^{2}=\bar{\sigma}^2=0.04,\eta
=0.39.$$

- <span> We showed earlier how to compute the value of a volatility swap
in terms of the Heston parameters.</span>

- <span> We can also easily compute the values of calls on variance in
terms of the Heston parameters. </span>

- <span>Given the values of the variance and the volatility swap, we can
also apply equation [(1)](#eq:BSvarcall) to value the same calls on
variance.</span>

### Constraints on volatility derivative pricing

- <span> We see that agreement is pretty close. 
    - And if agreement is so close in the Heston case where we believe the dynamics to be unrealistic, how much better should the agreement be when the underlying volatility dynamics are lognormal (the more realistic case)?


- <span> We conclude that the values of $T$-maturity volatility
derivatives are in practice tightly constrained by the prices of
expiration-$T$ European options. </span>

### Another potential recipe

- To make the argument even more explicit, we fit our preferred stochastic
volatility model (lognormal let’s say) to the prices of European
options, and we set the correlation $\rho$ to zero (this doesn’t affect
the value of volatility derivatives).



- <span>Now recall the following exact formula for the volatility swap under zero correlation:

    <a name="eq:volswap"></a>(2)
    $$\mathbb{E}\left[\sqrt{\langle x \rangle_T}\right]=
\sqrt{2\,\pi}\,\hat
c(0)+\\\quad\quad\quad\quad\quad\quad\sqrt{\frac{2}{\pi}}\,\int_0^\infty\,e^{k/2}\,I_1\left(\frac{k}{2}\right)\,c(k)\,dk
$$

  where $I_n(\cdot)$ represents a modified Bessel function of the first
kind.</span>

- Compute the value of the volatility swap using [(2)](#eq:volswap) and apply
our lognormal variance call option formula [(1)](#eq:BSvarcall), confident
that the result will be robust to the specific parameters.

    - The only constraint is that the fit to expiration-$T$ European prices
should be good.

### Options on volatility: More on model-independence

- <span> We have argued so far that given the prices of European options
of all strikes and expirations, the values of options on variance should
be tightly constrained.</span>


### A model-independent lower bound

- <span> [Dupire]<sup id="cite_ref-Dupire:2005" class="reference"><a href="#cite_note-Dupire:2005"><span>[</span>6<span>]</span></a></sup> shows how a model-independent lower bound for calls
on realized variance may be obtained by considering the delta hedging of
a portfolio of European call options. </span>

- <span>The idea is to hedge in business time defined as the timescale of
realized quadratic variation.</span>

    - <span>Options are rehedged each time quadratic variation increases by a
given amount.</span>

    - <span> Unlike conventional delta hedging in (atomic) clock time,
assuming no jumps, this hedging strategy guarantees profits on a short
option position if realized volatility is below implied volatility (the
volatility implied by the initial sale price).</span>

    - <span> Conversely the hedged option position is guaranteed to lose if
realized volatility is higher than implied at inception.</span>

- <span> Losses are insured against by buying an option on quadratic
variation. </span>


- <span> The hedger ends up short a portfolio of European options and long
an option on realized variance with only positive payoffs: this
portfolio must be worth at least zero and a model-independent lower
bound on the price of an option on variance follows. </span>

- <span> In the same work, Dupire shows (without formally proving) how
solutions to the Skorokhod Embedding problem can generate effective
upper and lower bounds on the price of an option on variance constrained
by known prices of European options.</span>

### A wrong conjecture

- <span>In [The Volatility Surface]<sup id="cite_ref-Gatheral" class="reference"><a href="#cite_note-Gatheral"><span>[</span>7<span>]</span></a></sup>, it was conjectured that the minimum possible value
of an option on variance is the one generated from a local volatility
model fitted to the volatility surface.</span>

    - <span>Options on variance have value even in a local volatility model
because realized variance depends on the realized path of the stock
price from inception to expiration.</span>

- <span> Given that local variance is a risk-neutral conditional
expectation of instantaneous variance, it seemed obvious that any other
model would generate extra fluctuations of the local volatility surface
relative to its initial state.</span>

### A counterexample due to Beiglböck et al.

Consider the following toy model from [Beiglböck et al.]<sup id="cite_ref-Beiglbock" class="reference"><a href="#cite_note-Beiglbock"><span>[</span>3<span>]</span></a></sup> where the volatility
path can be one of 

$$\begin{array}{cc}
v_+=\left\{
\begin{array}{ll}
  2   & \text{ if } t \in [0,1)  \\
  3   &  \text{ if } t \in [1,2) \\
  1   &  \text{ if } t \in [2,3] 
\end{array}
\right.
&
v_-=\left\{
\begin{array}{ll}
  2   & \text{ if } t \in [0,1)  \\
  1   &  \text{ if } t \in [1,2) \\
  3   &  \text{ if } t \in [2,3] 
\end{array}
\right.
\end{array}$$ 

depending on the toss of a coin.

Independent of the result of the coin-toss, the total variance
$$W_3=\int_0^3\,v_t\,dt=6$$ so ${\mathbb{E} }[(W_3-6)^+]=0$.

Now define

$$\Theta(x)={\partial}_T C_{BS};\,\Psi(x)=K^2\,{\partial}_{K,K}C_{BS};$$


Then local variance is given by

\begin{eqnarray*}
v_\ell(x,t)=2\,\frac{\partial_T C}{K^2\,\partial_{K,K}C}&=&2\,\frac{\Theta_+(x)+\Theta_-(x)}{\Psi_+(x)+\Psi_-(x)}\\
&=&\frac{v_+(t)\,\Psi_+(x)+v_-(t)\,\Psi_-(x)}{\Psi_+(x)+\Psi_-(x)}
\end{eqnarray*}

which is obviously a nonconstant function of $x$ and $t$.

One can verify with a Monte Carlo computation that


${\mathbb{E} }[(\tilde W_3-6)^+]\approx 0.026>0$.

In words, the value of the option on variance under local volatility is
greater than the value of the option on variance under stochastic
volatility in this case.

### Carr and Lee upper and lower bounds

- Given European options prices, [Carr and Lee]<sup id="cite_ref-CarrLee:2010" class="reference"><a href="#cite_note-CarrLee:2010"><span>[</span>4<span>]</span></a></sup> derive
model-independent upper and lower bounds for both spot- and forward-starting options on quadratic variation assuming only that the underlying is a positive continuous semimartingale. 


- Carr and Lee explicitly construct sub- and super-hedges.

    - The sub-hedge (and hence lower bound) is due to Dupire: Delta hedge a short call position in business time, making money if the realized variance is less than the
implied variance at the time of sale.

    - The super-hedge involves replicating in business time a double
knockout with a carefully chosen payoff at the barrier.

### Conclusion

- <span> Between these model-independent upper and lower bounds, it seems
reasonable to suppose that the fair value of an option on variance is
closest to that given by a stochastic volatility model.</span>

    - <span>It’s easy to reject local volatility.</span>

   

### The VIX index

- <span> In 2004, the CBOE listed futures on the VIX - an implied
volatility index.</span>

- <span> Originally, the VIX computation was designed to mimic the implied
volatility of an at-the-money 1 month option on the OEX index. It did
this by averaging volatilities from 8 options (puts and calls from the
closest to ATM strikes in the nearest and next to nearest
months).</span>

- <span>The CBOE changed the VIX computation: “CBOE is changing VIX to
provide a more precise and robust measure of expected market volatility
and to create a viable underlying index for tradable volatility
products.”</span>

### The VIX formula

Here is the revised VIX definition (converted to our notation) as specified
in the CBOE white paper:

<a name="eq:vixformula"></a>(3)
$$VIX^2=\frac{2}{T}\,\sum_i\,\frac{\Delta K_i}{K_i^2}\,
Q_i(K_i)\,-\,\frac{1}{T}\,\left[\frac{F}{K_0}-1\right]^2
$$

where $Q_i$ is the price of the out-of-the-money option with strike
$K_i$ and $K_0$ is the highest strike below the forward price $F$ .

We recognize [(3)](#eq:vixformula) as a straightforward discretization of
the log-strip and makes clear the reason why the CBOE implies that the
new index permits replication of volatility.

Specifically, (with obvious notation)

\begin{eqnarray*}
\frac{VIX^2\,T}{2}&=& \int _{0} ^F\, \frac{dK}{K^2} \,
P(K)\,+\,\int _F ^\infty \, \frac{dK}{K^2} \,C(K)\\
&=&  \int _{0} ^{K_0}\, \frac{dK}{K^2} \, P(K)\,+\,\int _{K_0}
^\infty \, \frac{dK}{K^2} \,C(K)+\int _{K_0}^F\, \frac{dK}{K^2}
\,\left( P(K)-C(K)\right)\\
&=:& \int _{0} ^{\infty}\, \frac{dK}{K^2} \, Q(K)\,+\int _{K_0}^F\,
\frac{dK}{K^2} \,\left(
K-F\right)\\
&\approx&\int _{0} ^{\infty}\, \frac{dK}{K^2} \,
Q(K)\,+\frac{1}{K_0^2}\,\int _{K_0}^F\, {dK} \,\left( K-F\right)\\
&=&\int _{0} ^{\infty}\, \frac{dK}{K^2} \,
Q(K)\,-\frac{1}{K_0^2}\,\frac{ \left(K_0-F\right)^2}{2}.
\end{eqnarray*}

One possible discretization of this last expression is
$$VIX^2=\frac{2}{T}\,\sum_i\,\frac{\Delta K_i}{K_i^2}\,
Q_i(K_i)\,-\,\frac{1}{T}\,\left[\frac{F}{K_0}-1\right]^2$$ as in the VIX
specification [(3)](#eq:vixformula).

### VIX futures and options

A time-$T$ VIX future is valued at time $t$ as


$${\mathbb{E} }_t\left[\sqrt{\frac 1 \Delta{\mathbb{E} }_T\left[\int_T^{T+\Delta}\,v_s\,ds\right]}\right]$$


where $\Delta$ is around one month (or $\Delta \approx 1/12$).

A VIX option expiring at time $T$ with strike $K_{VIX}$ is valued at
time $t$ as


$${\mathbb{E} }_t\left[\left(\sqrt{\frac 1 \Delta{\mathbb{E} }_T\left[\int_T^{T+\Delta}\,v_s\,ds\right]}-K_{VIX}\right)^+\right].$$

### Remarks on VIX futures and options

- <span>Historically, the options have been much more liquid and actively
traded than the futures.</span>


- <span>Both futures and synthetic futures (from put-call parity) used to
trade at a substantial premium to the forward-starting variance
swap.</span>

 - This arbitrage has come and gone over time.


- <span>Apparently, since the second quarter of 2013, there has been more vega traded in VIX than on SPX, which is now only the second biggest global market for volatility.</span>

   

### Forward starting variance swaps from VIX options

- <span>Note that we can span the payoff of a forward starting variance
swap ${\mathbb{E} }_t\left[\int_T^{T+\Delta}\,v_s\,ds\right]$ using VIX
options.</span>

- Recall the spanning formula: $$\mathbb{E} \left[ g(S_T)\right]
 =g(F)\,+\,\int _0 ^F \, dK \, \tilde
 P(K)\,g''(K)\,\\+\,\int _F ^\infty \, dK \,
  \tilde C(K)\,g''(K).$$

- In this case, $g(x)=x^2$ so
<p>
$${\mathbb{E} }_t\left[\int_T^{T+\Delta}\,v_s\,ds\right] = F_{VIX}^2  +  2 \int _0 ^{F_{VIX}} \, \tilde
 P(K) \, dK \\+ 2 \int _{F_{VIX}} ^\infty \, \tilde C(K)\,dK.$$

    - $F_{VIX}$ can be computed using put-call parity or observed directly as the VIX futures price.
<p> 
    - We need to interpolate and extrapolate out-of-the-money option prices to
get the <span>*convexity adjustment*</span>.

### Interpolation and extrapolation of VIX smiles


- We perform

    - Monotonic spline interpolation of mid-vols.

    - Extrapolation at constant level.
    
 
- Then we integrate the Black-Scholes formula with these vols.

<h3><img src="http://mfe.baruch.cuny.edu/wp-content/uploads/2014/12/BeijingVIX110915Interpolation2.png" align = "center" width=1000></h3> 

<a name="fig:spxLogRet"></a>
Figure 7: An example of a VIX smile with monotonic spline interpolation/
extrapolation.

```r
vix2 <- function(ivolData)function(slice){
  
  bidVols <- as.numeric(ivolData$Bid)
  askVols <- as.numeric(ivolData$Ask)
  expDates <- unique(ivolData$Texp)
  
  ###############################################################################################################
  # Interpolate and extrapolate vols for this slice at requested output points
  
  t <- expDates[slice]
  texp <- ivolData$Texp
  bidVol <- bidVols[texp==t]
  askVol <- askVols[texp==t]
  midVol <- (bidVol+askVol)/2
  f <- (ivolData$Fwd[texp==t])[1]
  k <- log(ivolData$Strike[texp==t]/f) # Plot vs log-strike
  include <- !is.na(bidVol)
  kmin <- min(k[include])
  kmax <- max(k[include])
  
  # Compute and store interpolated and extrapolated vols
  kIn <- k[!is.na(midVol)]
  volIn <- midVol[!is.na(midVol)]
  volInterp <- function(kout){
    if (kout < kmin){res <- midVol[which(k==kmin)] }
    else if( kout > kmax){res <- midVol[which(k==kmax)] }
    else res <- stinterp(x=kIn,y=volIn, kout)$y
    return(res)
  }
  vixVol <- function(x){sapply(x,volInterp)}
  # Now we use the vectorized function vixVol to compute the convexity adjustemnt
  cTilde <- function(y){exp(y)*BSFormula(1, exp(y), t, r=0, vixVol(y))}
  pTilde <- function(y){exp(y)*BSFormulaPut(1, exp(y), t, r=0, vixVol(y))}
  callIntegral <- integrate(cTilde,lower=0,upper=10)$value
  putIntegral <- integrate(pTilde,lower=-10,upper=0)$value
  res <- f^2*(1+2*(callIntegral+putIntegral))
  return(res)
}
```

### The VVIX index

- Recall that we may compute the fair value of quadratic variation of $\log S$ from the log-strip of SPX options:

$$
\E[\langle \log S \rangle_T] = -2\,\E[\log S_T].
$$

- We may also compute the fair value of quadratic variation of $\log VIX$ from the log-strip of VIX options:

$$
\E[\langle \log VIX \rangle_T] = -2\,\E[\log VIX_T].
$$

- The VVIX index is computed from the log-strip of VIX options in exactly the same way that VIX is computed from SPX options.

    - Roughly speaking, the VVIX index is the fair value of one-month quadratic variation of VIX.
    
    
- We then, again roughly speaking, also have a measure of the volatility of volatility.

### The fair value of VIX futures

As before, we have $F_{VIX} = \E[VIX]$ and

$$
\E[VIX^2] - (\E[VIX])^2 = \E[VIX^2] - F_{VIX}^2 = \mathop{Var}(VIX).
$$

Recall that $\E[VIX^2]$ may be computed from a constant strip of VIX options.  As for $\mathop{Var}(VIX)$, we have

$$
\mathop{Var}(VIX) \approx F_{VIX}^2\,VVIX^2\,T
$$

so 

$$
\E[VIX^2]  \approx F_{VIX}^2\,(1 + VVIX^2\,T)
$$

which gives yet another way of assessing the fair value of VIX futures.

### The rough Bergomi model

The rough Bergomi model may be written as

$$
\bea
\frac{dS_t}{S_t} &=& \sqrt{v_t}\,\left\{  \rho\,dW_t + \sqrt{1-\rho^2}\,dW^\perp_t \right\}\nonumber\\
v_u &=& \xi_t(u) \,\mathcal{E}\left(\eta\,\sqrt{2\,H}\, \int_t^u\,\frac{dW_s}{(u-s)^\gamma} \right)
\eea
$$
with $\gamma = \frac 12 - H$.

The forward variances $\xi_t(u) =  \eef{v_u}$ are (at least in principle) tradable and  observed in the market.  In forward variance form

$$
\frac{d \xi_t(u)}{\xi_t(u)} = \tilde \eta\,\frac{dW_t}{(u-t)^\gamma} 
$$

where $\tilde \eta = \eta\,\sqrt{2\,H}$.

- In particular, forward variances are lognormal.

### VIX options and futures under rough Bergomi

- The rough Bergomi model generates more or less flat VIX smiles.
    - Inconsistent with observed VIX smiles.
    
    
- Nevertheless, we can still try to compute the term structure of VIX futures to impute the rough Bergomi parameters $H$ and $\eta$.

### The distribution of VIX future payoffs


- Denote the terminal value of the VIX futures by $\sqrt{\zeta(T)}$.  Then, by definition,
<p>
$$
\zeta(T)= \frac{1}{\Delta}\,\int_T^{T+\Delta}\,\E[v_u|\cF_T]\,du = \frac{1}{\Delta}\,\int_T^{T+\Delta}\,\xi_T(u)\,du.
$$
<p>
where $\Delta$ is one month.


- Though a sum of lognormal random variables is not itself lognormal, if $\Delta$ is sufficiently small, $\zeta(T)$ should be approximately lognormal.




- The VIX payoff $\sqrt{\zeta(T)}$ should also be approximately lognormally distributed.

    -  The quality of this approximation was confirmed by <span>[Jacquier, Martini and Muguruza]<sup id="cite_ref-JMM" class="reference"><a href="#cite_note-JMM"><span>[</span>14<span>]</span></a></sup>.
    - In that case, the terminal distribution of $\zeta(T)$ is completely determined by $\eef{\zeta(T)}$ and $\var[\log \zeta(T)|\cF_t]$.
    

- Obviously
<p>
$$
\eef{\zeta(T)} = \frac{1}{\Delta}\,\int_T^{T+\Delta}\,\xi_t(u)\,du.
$$
<p>
    - Recall that forward variances $\xi_t(u)$ may be estimated from variance swaps which can themselves be proxied by the log-strip (see Chapter 11 of <span>[The Volatility Surface]<sup id="cite_ref-TVS" class="reference"><a href="#cite_note-TVS"><span>[</span>8<span>]</span></a></sup> again).
    
    - Alternatively they may be estimated from linear strips of VIX options.

### Approximating the conditional variance of $\zeta(T)$


- To estimate the conditional variance of $\zeta(T)$, we approximate the arithmetic mean by the geometric mean as follows:

<p>
$$
\zeta(T) \approx \exp\left\{\frac{1}{\Delta}\,\int_T^{T+\Delta}\,\E[\log v_u|\cF_T]\,du\right\}.
$$
<p>

Let $y_u = \log v_u$ and recall that $\gamma = \frac12 - H$.  Apart from $\mathcal{F}_t$ measurable terms (abbreviated as ``drift''), we have

$$
\beas
  \int_T^{T+\Delta} E[y_u|\mathcal{F}_T]\, du &=& \eta \sqrt{2H} \int_t^T
  \frac{dW_s}{(u-s)^\gamma} du + \text{drift}\\
  &=& \eta \sqrt{2H} \int_t^T
  \int_T^{T+\Delta}  \frac{du}{(u-s)^\gamma} dW_s + \text{drift} \\
  &=& \eta \frac{\sqrt{2H}}{1-\gamma} \int_t^T \left[ (T+\Delta-s)^{1-\gamma} - (T-s)^{1-\gamma}
  \right] dW_s + \text{drift}.
\eeas
$$

This gives

$$
\beas
  \var[\log \zeta(T) | \mathcal{F}_t] &\approx& \frac{\eta^2
    }{\Delta^2} \,\frac{2 H}{(H+1/2)^2}\,\int_t^T \left[(T+\Delta-s)^{1/2+H} - (T-s)^{1/2+H}
  \right]^2 ds\\
  &=& \eta^2\,(T-t)^{2\,H}\,f^H\left(\frac{\Delta}{T-t}\right)
\eeas
$$

where

$$
f^H(\theta) = \frac{2 H}{(H+1/2)^2}\,\frac{1}{\theta^2}\, \int_0^1\,\left[(1+\theta-x)^{1/2+H} - (1-x)^{1/2+H}
  \right]^2 dx.
$$
<p>

### The Double Mean Reverting (DMR) model

One of the earliest attempts to jointly model SPX and VIX was the DMR<sup id="cite_ref-GatheralBachelier" class="reference"><a href="#cite_note-GatheralBachelier"><span>[</span>8<span>]</span></a></sup>.</span> model with dynamics 

<a name="eq:doubleCEVdymamics"></a>(4)
$$
\beas
\frac{dS}{S}&=&\sqrt{v}\,dW\nonumber\\
dv&=&-\kappa\,(v-v')\,dt+\eta_1\,{v}^\alpha\,dZ_1\nonumber\\
dv'&=&-c\,(v'-z_3)\,dt+\eta_2\,{v'}^\beta\,dZ_2
\eeas
$$
<p><p>
for any choice of $\alpha,\, \beta \in [1/2,1]$.  The short term variance level $v$ reverts to a moving level $v'$ at rate $\kappa$. $v'$ reverts to the long-term level $z_3$ at the slower rate $c<\kappa$.
    
- We call the case $\alpha=\beta=1/2$ *Double Heston*,
- the case $\alpha=\beta=1$ *Double Lognormal*,
- and the general case *DMR*.

### Rough volatility models

- Though we don't show it here, it's clear that rough Heston will generate the same unreasonable downward sloping VIX smiles as the classical Heston model.


- The rough Bergomi model generates almost flat VIX smiles so cannot fit VIX smiles.


- Bergomi addresses this issue (in his 2-factor model) by modeling instantaneous variance as a linear combination of two Ornstein-Uhlenback (OU) processes with different mean reversion rate rather than as lognormal.


- [Horvath, Jacquier and Tankov]<sup id="cite_ref-Horvath" class="reference"><a href="#cite_note-Horvath">[13]</a></sup> implement a rough version of Bergomi's idea by modeling instantaneous variance as a sum of Riemann-Liouville processes.
   

### Martingale optimal transport

- The idea is to bound the expected value of a claim over all martingale measures that have a given marginal at maturity.
    - Typically, super- and sub-replicating portfolios are also given.


- A 2015 paper by [De Marco and Henry-Labordère]<sup id="cite_ref-PHL" class="reference"><a href="#cite_note-PHL">[5]</a></sup> applies martingale optimal transport (MOT) to SPX and VIX options.
    - They claim that on July 12, 2013, "market prices of VIX options are well above our analytical upper bound for $K/VIX >1$."
    
    
- Their conclusion is therefore that VIX and SPX cannot be jointly modeled.

### Upper bound on VIX futures

[De Marco and Henry-Labordère]<sup id="cite_ref-PHL" class="reference"><a href="#cite_note-PHL">[5]</a></sup>  present the following *nice-looking* argument showing that the VIX future should be bounded above by its value in a local volatility model.

- For $\Delta/T$ sufficiently small
$$
\beas
VIX_t(T) &=& \eef{\sqrt{\frac{1}{\Delta}\int_T^{T+\Delta}\,\xi_T(s)\,ds}}\\
&\approx& \eef {\sqrt{V_T}} \\
&=& \eef{\ee{\sqrt{V_T}|S_T}} \\
&\leq& \eef {\sqrt{\ee{V_T|S_T}}} \text{ (Jensen)}\\
&=& \eef {\sqrt{v_\ell(S_T,T)}}\\
&\approx & VIX^{LV}_t(T).
\eeas 
$$



- According to the authors "market prices of VIX futures typically lie in practice above prices given by the local volatility model", *i.e.* above the upper bound.
    - Very reminiscent of my earlier wrong conjecture.


- [Acciaio and Guyon]<sup id="cite_ref-Acciaio" class="reference"><a href="#cite_note-Acciaio">[1]</a></sup> show that the argument of De Marco and Henry-Labordère is not correct in general.
    - Using an argument similar to that of Beiglböck et al., they give a counterexample to the claim that the value of VIX futures should be maximized under local volatlity.

### Later developments



- In 2020, [Guyon]<sup id="cite_ref-Guyon" class="reference"><a href="#cite_note-Guyon">[11]</a></sup> showed using MOT methods that it is possible to calibrate models to SPX and VIX simultaneously at discrete times.
    - However, he conjectured that it is not possible to find models with continuous sample paths that jointly fit these two volatility surfaces.
    
    
    
- Later on, it was shown numerically that the [Quadratic Rough Heston]<sup id="cite_ref-QRH" class="reference"><a href="#cite_note-QRH">[9]</a></sup> model could jointly fit both SPX and VIX smiles.


- Still later, [Guo et al.]<sup id="cite_ref-Guo" class="reference"><a href="#cite_note-Guo">[10]</a></sup> showed how to calibrate a (continuous time) local stochastic volatility model to SPX and VIX simultaneously using MOT methods by solving an HJB equation numerically.
    - So it seems that models with continuous paths can jointly fit SPX and VIX.
    - But the calibration procedure of Guo et al. is extremely complicated.



### More recently

- More recently, in a wonderful paper, [Rømer]<sup id="cite_ref-Romer" class="reference"><a href="#cite_note-Romer">[16]</a></sup> jointly calibrates a number of models to SPX and VIX using machine learning methods.
    - The dataset is SPX and VIX options from May 2004 to May 2019.

### Shifted mixture rough Bergomi model (S-2F-RB)

As usual,

$$
    \frac{dS_t}{S_t}=\sqrt{V_t}\,dZ_t.
$$

Let $\xi_t(u) = \eet{V_u}$. Then

$$
\beas
V_t &=& \zeta_0(t)\,\left\{\mu\,X_{1,t} +(1-\mu)\,X_{2,t}\right\}\\
X_{i,t}&=& \cE\left(\eta_i\,\delta_i\,\left[\theta_i\,Y_{1,t}+(1-\theta_i)\,Y_{2,t}\right]\right)\\
Y_{i,t}&=& \int_0^t\,K_i(t-s)\,dW_{i,s}, \quad i=1,2,\,t \geq 0.
\eeas
$$

In particular, note that $\xi_0(t)= \zeta_0(t)+c$.  $K(\tau) = \tau^{\alpha-1}\,e^{-\lambda t}$.

### Quadratic rough Heston (QRH)

$$
\beas
V_t &=& a\,(Z_t+b)^2+c\\
Z_t&=& \zeta_{0}(t)+\int_0^t\,\eta\,K(t-s)\,\sqrt{V_s}\,dW_{s}.
\eeas
$$



- This is a simple generalization of the [Quadratic Rough Heston]<sup id="cite_ref-QRH" class="reference"><a href="#cite_note-QRH">[11]</a></sup> model.

### Hyperbolic version of quadratic rough Heston (S-M-2F-QRHyp)

$$
\beas
V_t &=& \mu\,X_{1,t} +(1-\mu)\,X_{2,t}+c\\
X_{i,t}&=& f_H(Z^2_{i,t}-d_i)-f_H(-d_i)\\
Z_{i,t} & =& \eta_i\,\delta_i\,\left[\theta_i\,Y_{1,t}+(1-\theta_i)\,Y_{2,t}\right]\\
Y_{i,t}&=& \zeta_{i,0}(t)+\int_0^t\,K_i(t-s)\,dW_{i,s}, \quad i=1,2,\,t \geq 0.
\eeas
$$

where $f_H(x) = x+\sqrt{x^2+1}$.

- The hyperbolic transformation $f_H$, originally suggested by [Jäckel and Kahl]<sup id="cite_ref-KJ" class="reference"><a href="#cite_note-KJ">[15]</a></sup>, has the same behavior as $\exp$ for small $x$ but grows much more slowly for large $x$.
    - This has the effect of reducing the growth rate of moments relative to typical lognormal models.

### Rømer conclusions


- In terms of volatility prediction, rough Bergomi beats rough Heston.  Both rough models do much better than classical Heston.



- At least two stochastic factors are required to model the variance process.



- There is a pronounced term structure of smile effect where short-dated SPX smiles are more symmetric than longer-dated ones.


- The tails associated with lognormal models are too fat.
    - Quadratic or hyperbolic models are preferred.
    
  

- To quote from Rømer:
<p><p>
Moreover, we were for several reasons unable to calibrate, across volatility levels, jointly to SPX and VIX options with more general models based on the log-normal distribution. We therefore considered several alternatives, including a two- factor volatility model based on the quadratic and hyperbolic transformations and which is driven by two ordinary Ornstein- Uhlenbeck processes. We demonstrated that this model can calibrate almost perfectly to SPX options and very decently to the joint problem, also across volatility levels. From this we conclude that the SPX and VIX options markets can largely be reconciled with classical two-factor volatility, all without roughness and jumps.


### Most recently

- Most recently, in a  very impactful paper, [Guyon and Lekeufack]<sup id="cite_ref-GL" class="reference"><a href="#cite_note-GL">[12]</a></sup> find a simple purely path-dependent Markovian model that fits both SPX and VIX.
    - At least on June 2, 2021.
    
    
- This is obviously consistent with Rømer's findings.

### The Guyon-Lekeufack (GL) model

This purely path-dependent model reads:

$$
\beas
\frac{dS_t}{S_t } &=& -\sigma_t\,dW_t\nonumber\\
\sigma_u &=& \beta_0 + \beta_1\, R_{1,u} + \beta_2\,\sqrt{R_{2,u}}\nonumber\\
R_{1,u} &=& \int_{-\infty}^u\,\kappa_1(u-s)\,\sigma_s\,dW_s\nonumber\\
R_{2,u} &=& \int_{-\infty}^u\,\kappa_2(u-s)\,\sigma_s^2\,ds.
\eeas
$$

### The 4FMPDV version of the GL model


- Guyon and Lekeufack find that shifted power-law kernels of the form $\kappa(\tau)=
(\delta+\tau)^{\alpha-1}$ describe the data best.

    - Such kernels are technically not rough but they are rough-like.


- The Markovian 4FMPDV version of the GL model has kernels
    <p><p>
$$
\beas
\kappa_1(\tau) &=& (1 - \theta_1)\,\lambda_{1,0}\,e^{-\lambda_{1,0}\,\tau} + \theta_1\,\lambda_{1,1}\,e^{-\lambda_{1,1}\,\tau}\\
\kappa_2(\tau) &=&(1 - \theta_2)\,\lambda_{2,0}\,e^{-\lambda_{2,0}\,\tau} + \theta_2\,\lambda_{2,1}\,e^{-\lambda_{2,1}\,\tau}.
\eeas
$$


- Because the 4FMPDV model is Markovian, it is easy to simulate.


- The initial values $R_{1,t}$ and $R_{2,t}$ are computed from historical returns and historical squared returns.
    - We can see why this paper is popular with over 7,000 downloads.

### ETFs according to Avellaneda and Zhang

Following Avellaneda and Zhang, denote the spot price of the underlying by
$S_t$, the price of the leveraged ETF (LETF) by $L_t$, and leverage
ratio by $\beta$.


Assume that $$\frac {dS_t}{S_t} = \mu_t\,dt+\sigma_t\,dW_t.$$

Then
assuming zero interest rates, dividends and borrowing costs, we have

$$\frac {dL_t}{L_t} = \beta\,\frac {dS_t}{S_t}$$

### Leveraged ETF valuation

Applying Itô’s Lemma,
$$d\log S_t=\frac {dS_t}{S_t} -\frac 12\,\sigma_t^2\,dt$$ and
$$d\log L_t=\frac {dL_t}{L_t} -\frac 12\,\beta^2\,\sigma_t^2\,dt=\beta\,\frac {dS_t}{S_t} -\frac 12\,\beta^2\,\sigma_t^2\,dt$$
Then

<a name="eq:LS"></a>(7)

$$
d\log L_t-\beta\,d\log S_t= -\frac 12\,(\beta^2-\beta)\,\sigma_t^2\,dt
$$

Integrating [(7)](#eq:LS) gives

<a name="eq:LETF"></a>(8)
$$
\frac{L_t}{L_0}=\left(\frac{S_t}{S_0}\right)^\beta\,\exp\left\{\frac 12\,(\beta-\beta^2)\,\int_0^t\,\sigma_s^2\,ds\right\}
$$

### Volatility exposure of leveraged ETF

- We see from [(7)](#eq:LS) and [(8)](#eq:LETF) that long positions in leveraged ETFs are effectively short volatility.

- For example:

    - If $\beta=2$, 
    
    $$d\log L_t-\beta\,d\log S_t= -\sigma_t^2\,dt.$$

    - Likewise, if $\beta=-1$,

$$d\log L_t-\beta\,d\log S_t= -\sigma_t^2\,dt.$$
- This is because the LETF has to buy the underlying when it goes up and
sell it when it goes down.

### Rebalancing

- Wlog, suppose the starting value of the LETF is $N_0=\$1$.

- At the close of day 1, the notional is
$$V_1=1+\beta\,\frac{\Delta S_1}{S_0}$$

- After rebalancing, at the close of day 2, the value is

  <a name="eq:rebal"></a>(9)
  $$
  V_2=V_1\,\left(1+\beta\,\frac{\Delta S_2}{S_1}\right)=\left(1+\beta\,\frac{\Delta S_2}{S_1}\right)\,\left(1+\beta\,\frac{\Delta S_1}{S_0}\right)
  $$

- On the other hand if there is no rebalancing the value would be

  <a name="eq:norebal"></a>(10)
  $$
  V_2'=\left(1+\beta\,\frac{\Delta S_1+\Delta S_2}{S_0}\right)
  $$
  dollars worth of underlying.

- Subtracting [(10)](#eq:norebal) from [(9)](#eq:rebal), we obtain
<p>
$$
\beas
  V_2-V_2'&=&\beta\,\left\{\frac{\Delta S_2}{S_1}-\frac{\Delta S_2}{S_0}+\beta\, \frac{\Delta S_2}{S_1}\,\frac{\Delta S_1}{S_0}  \right\}\\
  &=&\beta\,(\beta-1)\, \frac{\Delta S_1}{S_0} \,\frac{\Delta S_2}{S_1}
\eeas
$$
<p>
  from which we conclude that the rebalancing at the end of day 1
involved buying 
  $$\beta\,(\beta-1)\, \frac{\Delta S_1}{S_0}$$
  dollars worth of the underlying.


- Being long a leveraged ETF is therefore like being short gamma: you effectively buy(sell) the underlying
if the price rises(falls).
    - In aggregate therefore, leveraged ETFs exarcebate market moves.

Merge price histories into one dataset and label columns

```r
portfolio.prices <- as.xts(merge(Cl(SVXY),Cl(VXX),1)) # Add column for cash
portfolio.prices <- portfolio.prices["20180429::"] # Last five years only
names(portfolio.prices) <- c("SVXY","VXX","Cash")

portfolio.returns <-na.omit(ROC(portfolio.prices,1,"discrete"))
```

```r
plot(Cl(SVXY),main="",col="blue")
```

<a name="fig:vxxxiv.png"></a>Figure 20: SVXY performance.

### Rebalance the portfolio according to 50% SVXY, 50% cash

```r
w1 <- c(.5,0,0.5) # 50% SVXY, 50% cash
portfolio.rebal1 <- Return.portfolio(portfolio.returns,
                                         rebalance_on="weeks",
                                         weights=w1,wealth.index=TRUE,verbose=TRUE)
 
```

### Rebalance the portfolio according to short 25% VXX, 125% cash

```r
w2 <- c(1.25,0,-.25)# -25% VXX 125% cash 
portfolio.rebal2 <-Return.portfolio(portfolio.returns,
                                         rebalance_on="weeks",
                                         weights=w2,wealth.index=TRUE,verbose=TRUE)
 
```

### Rebalance the portfolio according to 50% SVXY, 25% VXX, 25% cash

```r
w3 <- c(.25,.5,.25) # 25% cash, 50% SVXY, long 25% VXX 
portfolio.rebal3 <- Return.portfolio(portfolio.returns,
                                         rebalance_on="weeks",
                                         weights=w3,wealth.index=TRUE,verbose=TRUE)
 
```

### Buy and hold SVXY (no rebalancing)

```r
w4 <- c(1,0,0) # 100% SVXY
portfolio.bh <-Return.portfolio(portfolio.returns,
                                         weights=w4,wealth.index=TRUE,verbose=TRUE)
 
```

Merge portfolio returns into one dataset and label columns

```r
portfolios <-cbind(portfolio.rebal1$returns,portfolio.rebal2$returns,portfolio.rebal3$returns,portfolio.bh$returns,0)
colnames(portfolios) <-c("50% SVXY, 50% cash","Short 25% VXX, 125% cash","50% SVXY, 25% VXX, 25% cash","Buy and hold SVXY","Cash")
```
